# Introdução ao Aprendizado de Máquina

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error, r2_score
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
```

```python
_, ax = plt.subplots()

scatter = ax.scatter(iris.data[:, 1], iris.data[:, 2], c = iris.target)

ax.set(xlabel = iris.feature_names[1], ylabel = iris.feature_names[2])

ax.legend(scatter.legend_elements()[0], iris.target_names, loc = "upper right", title = "Classes")
```

```python
ax2 = sns.scatterplot(data = iris, x = iris.data[:, 1], y = iris.data[:, 2], hue = iris.target)

h, labels = ax2.get_legend_handles_labels()

ax2.legend(h, iris.target_names, loc='upper right')
```

## Construção de um classificador

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.33, random_state = 42)
```

```python
modelo = KNeighborsClassifier(n_neighbors = 5)
```

```python
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

## Construção de um modelo de regressão

```python
X, y = datasets.load_diabetes(return_X_y = True)
```

```python
y
```

```python
X.shape
```

```python
X1 = X[:, [2]]
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X1, y, test_size = 0.3, shuffle = False)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train2, y_train2)
```

```python
y_pred2 = regressor.predict(X_test2)
```

```python
print(mean_squared_error(y_test2, y_pred2))
print(r2_score(y_test2, y_pred2))
```

```python
_, ax = plt.subplots()

ax.scatter(X_test2, y_test2)

ax.plot(X_test2, y_pred2, linewidth = 2, color="red")
```
